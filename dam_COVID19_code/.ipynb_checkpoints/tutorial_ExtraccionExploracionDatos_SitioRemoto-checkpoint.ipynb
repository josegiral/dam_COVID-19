Para extraer los archivos csv a un cuadro de datos en pandas, es necesario saber el tipo de codificación de los caracteres y si están comprimidos, el tipo de compresión. Un cuadro de datos es un objeto (definición de objeto en python [aquí](https://docs.python.org/3/tutorial/classes.html)) en python cuyos datos son la información del archivo csv, y los métodos son funciones u otras secuencias de operaciones que se pueden realizar sobre los datos.  

Por ejemplo, 

```python
data=pd.read_csv(sitio+archCasos, compression=None, encoding='latin-1')
```

En la medida de lo posible, siempre es conveniente construir funciones para poder llevar a cabo procesos que pudiera interesarnos repetir. En este caso, una función para abrir archivos con pandas con la que podemos repetir el proceso de apertura para varios archivos. Por ejemplo, 

```python
def openCSV_DB(path, comp=None,enc='latin-1'):
    data=pd.read_csv(path, compression=comp,encoding=enc)
    print('Data obtained from %s'%path)
    return data
```

Si queremos abrir otros tres archivos, podemos usar la función openCSV_DB. 

```python
sitio='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/'
archCasos='time_series_covid19_confirmed_global.csv'
archFatal='time_series_covid19_deaths_global.csv'
archRecup='time_series_covid19_recovered_global.csv'

compr=None; codif='latin-1'
casos = openCSV_DB(sitio+archCasos,compr,codif)
# Los argumentos de una función pueden aparecer en orden distinto al orden en que fueron introducidos
recup= openCSV_DB(enc=codif,path=sitio+archRecov,comp=compr)
# Usando los valores para los argumentos que están previamente especificados
fatal= openCSV_DB(path=sitio+archFatal,comp=compr)
```

*Cambio en el tipo de compresión.*  Si los archivos csv están comprimidos, también es posible abrirlos con read_csv hay que cambiar el argumento de compresión en la función read_csv.

Por último, construyamos un diccionario de datos sólo con los casos, recuperaciones y muertes de un país, pero ahora en arreglos de numpy (para poder hacer operaciones y análisis con ellos).

El caso de China es útil porque hay datos para varias provincias. 

```python
cc = 'China'
China = dict()
#
idx = casos['Country/Region'].isin([cc])
China['fechas'] = casos.columns[4:].to_numpy()
China['provincias'] = casos[casos['Country/Region'].isin([cc])].iloc[:,0].to_numpy()
print(China['provincias'])
China['aCasos_prov']= casos[casos['Country/Region'].isin([cc])].iloc[:,4:].to_numpy()
China['aFatal_prov']= deaths[deaths['Country/Region'].isin([cc])].iloc[:,4:].to_numpy()
China['aRecup_prov']= recup[recup['Country/Region'].isin([cc])].iloc[:,4:].to_numpy()
mm=3
print('Las dimensiones de la entrada de aCasos_prov son ',China['aCasos_prov'].shape)
print('Casos en cada provincia (un arreglo por provincia)\n',China['aCasos_prov'])
print('Casos en la provincia de %s \n'%China['provincias'][mm],China['aCasos_prov'][0])
```

La ventaja es que ahora los datos pueden ser manipulados más fácilmente y además se pueden utilizar todas las funciones de numpy sin tener que interactuar con el módulo pandas. Un ejemplo trivial es que podemos obtener los datos de todo el país sumando sobre la primera dimensión del arreglo de casos (lo mismo para las recuperaciones y las defunciones). 

```python
# Datos de todo el país
China['aCasos']= China['aCasos_prov'].sum(0)
China['aRecup']= China['aRecup_prov'].sum(0)
China['aFatal']= China['aFatal_prov'].sum(0)
```

Ahora grafiquemos los datos desde este nuevo formato, pero primero obtengamos prevalencia e incidencia sin normalizar con respecto al tamaño poblacional para graficar los datos de una manera distinta. 

```python
China['prevalencia']= China['aCasos'] - (China['aRecup'] + China['aFatal'])
China['d_prev'] = np.zeros(len(China['prevalencia']))
China['d_prev'][1:] = China['prevalencia'][1:]-China['prevalencia'][:-1]
```

```python
nD=len(China['fechas']); iD = np.arange(0,nD,7)
#
g1 = gr.figure(figsize=(15,5)); gr.ioff() # Modo interactivo apagado. Todos los elementos de la gráfica son puestos en el objeto gráfico al mismo tiempo en lugar de reescribir la gráfica varias veces.
tx = g1.add_subplot(1,1,1); 
for yy in range(len(China['provincias'])):
    tx.plot(China['aCasos_prov'][yy],'.',label= '%s'%China['provincias'][yy])
# Los índices del eje horizontal son 0,1,...,n-1  donde n es el número de puntos a graficar 
# Ajustar los ticks a que sean sólo un subconjunto de esos índices en saltos de 7 
tx.set_xticks(iD) 
tx.set_xticklabels(China['fechas'][iD], rotation=90,fontsize=8)
tx.legend(loc='upper right',ncol=3)
tx.set_ylim(0,2000)
#
g2 = gr.figure(figsize=(15,5)); gr.ioff() 
filas =1; cols=2
ax = g2.add_subplot(filas,cols,1)
bx = g2.add_subplot(filas,cols,2)
ax.plot(China['prevalencia'], '.',lw=1,label= 'Prevalencia de casos') #estos no habían sido transpuestos
ax.plot(China['aCasos'], '--',lw=1,label= 'Casos acumulados')
ax.plot(China['aRecup']+China['aFatal'], 'k--',lw=1,label= 'Recuperaciones y muertes')
ax.plot(China['aRecup'], '-',color='green',lw=1,label= 'Recuperaciones')
ax.plot(China['aFatal'], 'k.', ms=2, label= 'Muertes')
ax.legend(loc='upper left')
ax.set_xticks(iD)
ax.set_xticklabels(China['fechas'][iD], rotation=90,fontsize=8)
#
bx.plot(China['d_prev'],China['prevalencia'],'.')
bx.plot([0,0],[China['prevalencia'].min(),China['prevalencia'].max()],'k:')
bx.set_ylabel('Prevalencia'); bx.set_xlabel('Cambio en la prevalencia')
bx.set_title('Retrato de fase de la prevalencia')
gr.show(p)
```
